```python
# List examples
numbers = [1,2,3,4,5]
fruits = ['apples', 'oranges', 'banana', 'pears']
fruits.append('coconuts')
fruits.remove('banana')

print(fruits[1])

# Insert into position
fruits.insert(2, 'strawbarries')

# Remove with pop
fruits.pop(-3)

# reverse the fruits
fruits.reverse()

# Sort
fruits.sort()

#Reverse sort
fruits.sort(reverse=True)

# Change values
fruits[0] = 'grapes'

print(fruits)
```

```python
#Tuples
#A tuple is a collection which is ordered and unchangeable. Allows duplicate members.

# Create tuple
friuits = ('Apples', 'Oranges', 'Grapes')

#single value needs trailing comma
fruits2 = ('Apples',)

#Get value
print(fruits[1])

fruits[0] = 'Pineaples'

print(fruits)
```

```python
#Sets
# A set is a collection which is unordered and unindexed. No duplicate memebers.

#create set
fruits_set = {'Apples', 'Oranges', 'Mangoes'}

#check if in set
print('Apples' in fruits_set)

# Add to set
fruits_set.add('Grapes')

#Remove from set
fruits_set.remove('Grapes')

print(fruits_set)
```
